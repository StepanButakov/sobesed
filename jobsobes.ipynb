### 1 задание

```python
def is_isomorphic(s, t):
    if len(s) != len(t):
        return False
    
    return len(set(zip(s, t))) == len(set(s)) == len(set(t))


s = 'paper'
t = 'title'
print(is_isomorphic(s, t)) 
```

#### Решение имеет временную сложность O(n) и пространственную сложность O(n), где n — длина строк.

### 2 задание

```python
def missing_number(nums):
    return list(set([i for i in range(min(nums), max(nums)+1)]) - set(nums))[0]
```

```python
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
print(missing_number(nums))
```

#### Решение имеет временную сложность O(n) для создания множеств и списков, и пространственную сложность O(n) для хранения всех чисел в диапазоне.

### 3 задание

```python
def simple(num):
    count = 0
    for i in range(1, num+1):
        if num % i == 0:
            count += 1
    if count == 2:
        return 'simple'
    else:
        return 'error'
```

```python
def prime_factors(num):
    A = []
    for i in range(2, num+1):
        if simple(i) == 'simple':
            while num % i == 0:
                num /= i
                A.append(i)
    return A
```

```python
print(prime_factors(56))
```

#### Решение имеет временную сложность O($n^{2}$) из-за вложенных циклов. Пространственная сложность O(k), k — количество простых множителей.
